```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

### Filtering and Obtaining data for the type "Movie" from the netflix dataframe

```python
movies_df = netflix_df[netflix_df["type"]=="Movie"]
movies_df.head(5)
```

### Restricting the movie data to only include data from the release_year 1990 till 1999

```python
import numpy as np
new_df = movies_df[np.logical_and(movies_df["release_year"] > 1989, movies_df["release_year"] < 2000)]
```

```python
new_df.head(5)
```

```python
new_df.shape
```

### Obtaining a dictionary of duration as keys and frequency as values

```python
rate = {}
for data in new_df["duration"]:
    rate[data] = rate.get(data, 0) +1
print(rate)
```

### Obtaining the highest value of the frequency and the corresponding key

```python
highest_value = max(rate.values())
highest_value
```

```python
duration = max(rate, key=rate.get)
duration
```

### Let's find the number of short action movies, i.e movies less than 90mins

```python
a = new_df[np.logical_and(new_df["duration"] < 90, new_df["genre"]=="Action")] 
```
